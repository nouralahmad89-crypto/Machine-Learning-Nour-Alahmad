### Support vektor Mashine (SVM):

SVM inte bara hittar en linje, utan den linje som ligger så långt bort som möjligt från punkterna på båda sidor. Det är alltså maximum margin-idén. Support vectors är de punkter som ligger närmast gränsen och som bestämmer var linjen hamna.

vi använder **KMeans** först för att skapa labels , KMeans delar upp datan i 2 grupper
dessa grupper blir våra labels ,nu kan vi träna en klassificeringsmodell

#### Vad SVM egentligen gör (Maximum Margin)?

Vi vill ha den bästa linjen, som längst bort från datapunkterna.

margin = avstånd från linjen till närmaste punkter.--> **SVM** vill maximera detta avstånd.

**stor margin = mer robust , mindre känslig för brus , bättre generalisering**

de punkter nära gränsen är support vectors , bara dessa punkter påverkar modellen.

#### Parameter: C

detta styr hur strikt modellen är:

- Hög C (hård) : försöker klassificera alla rätt , smal margin , risk för overfitting.
- Låg C (mjuk) : accepterar några fel , bred margin , mer robust.

C bestämmer hur mycket modellen bryr sig om fel jämfört med att hålla en stor margin.

#### Kernel Trick (hur SVM klarar svår data):

Vad är en kernel? En kernel bestämmer formen på beslutsgränsen

linear → rak linje
andra → böjda/komplexa gränser

##### 🔹 Linear Kernel
- Form: Rak linje
- Användning: Linjärt separerbar data
- Fördelar: Enkel, snabb, lätt att tolka
- Nackdel: Klarar inte komplex data

 Minnesregel: *"rak linje räcker"*

---

##### 🔹 Polynomial Kernel
- Form: Böjd kurva (polynom)
- Parameter: `degree`
- Användning: Lite komplex data
- Fördelar: Mer flexibel än linear
- Nackdel: Kan bli överkomplex

 Minnesregel: *"kontrollerad böjning"*

---

##### 🔹 RBF Kernel (viktigast)
- Form: Mycket flexibel (cirklar, blobbar)
- Användning: Icke-linjär data
- Fördelar: Väldigt kraftfull, funkar i många fall
- Nackdel: Kräver tuning

 Minnesregel: *"formar sig efter datan"*

---

##### 🔹 Sigmoid Kernel
- Form: Liknar neural network
- Användning: Sällan
- Nackdel: Instabil, ofta sämre än RBF

 Minnesregel: *"finns, men används sällan"*

---

##### Jämförelse

| Kernel      | Form         | Flexibilitet | När används        |
|-------------|-------------|--------------|--------------------|
| Linear      | Rak         | Låg          | Enkel data         |
| Polynomial  | Böjd        | Medium       | Lite komplex data  |
| RBF         | Fri form    | Hög          | Komplex data       |
| Sigmoid     | Speciell    | Varierar     | Sällan             |

---

Kernel trick:gör data separerbar, Hur: projicerar till hög dimension utan att räkna allt.

om vi bara använder linje → SVM är inte så användbar.

- linjär data → linjär modell
- icke-linjär data → kernel SVM

SVM kan nu:

- ✔️ lösa icke-linjära problem
- ✔️ skapa komplexa gränser
- ✔️ ändå vara effektiv

SVM handlar om:

- hitta bästa gräns (max margin)
- justera strikthet (C)
- välja rätt kernel
