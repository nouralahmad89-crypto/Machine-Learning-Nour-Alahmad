# **Machine Learning Landscape**

Maskininlärning kan delas in i fyra huvuddimensioner.  
När du lär dig en algoritm ska du kunna placera den i varje kategori.

---

## **Inlärningsparadigm (Hur lär sig modellen?)**

### 🔵 Supervised Learning
- Har etiketter (facit)
- Modellen lär sig genom att minimera fel.

Du visar exempel med facit.
- Bild + “det här är en hund”
- Bild + “det här är en katt”

Roboten lär sig koppla in → rätt svar.

---

### 🟡 Unsupervised Learning
- Inga etiketter
- Du har inga etiketter alls (inget facit). Modellen försöker bara hitta struktur
- Exempel:
   sortera saker i grupper som “liknar varandra” (klustring)

---

### 🟠 Semi-Supervised Learning
- Lite etiketter
- Mycket oetiketterad data
- Används när etiketter är dyra

---

### 🟣 Self-Supervised Learning
- Maskar delar av input
- Förutsäger saknade delar
- Du har inte facit från början, men du kan skapa en “uppgift” av datan.
- Exempel:
  - GPT (maskar ord)
  - täck över en del av bilden/texten och be modellen gissa det som saknas.

---

### 🔴 Reinforcement Learning
- Lär sig via belöning
- Interagerar med miljö
- Action → Reward → Uppdatering
- Exempel:
  Gör rätt → får belöning
  Gör fel → får mindre belöning/straff

Modellen lär sig genom att prova om och om igen och bli bättre över tid.

---

## **Typ av Uppgift (Task)**

### Regression
- Förutsäger numeriskt värde
- Exempel:
  - Pris
  - Temperatur
  - Tid

---

###  Klassificering
- Förutsäger kategori
- Exempel:
  - Spam
  - Sjukdom
  - Bildklassificering

---

###  Klustring
- Hittar grupper
- Ofta Unsupervised

---

### Dimensionality Reduction
- Minskar antal features
- Exempel:
  - trycka ihop info smart (som att sammanfatta)
  - Kompression

---

###  Anomaly Detection
- Hittar ovanliga mönster
- Exempel:
  - Kreditkortsbedrägeri

---

###  Novelty Detection
- Hittar något nytt (positivt ovanligt)

---

## **Träningssätt**

### 🔵 Batch Learning
- Tränas en gång
- Deployas
- Uppdateras inte kontinuerligt

---

### 🟡 Online Learning
- Uppdateras kontinuerligt
- Exempel:
  - Rekommendationssystem
  - Realtidsdata

---

##  **Modelltyp**

### 🔵 Instance-Based
- Lagrar data
- Jämför nya datapunkter med gamla
- Ingen riktig generalisering
- Exempel:
  - KNN

---

### 🟡 Model-Based
- Lär parametrar
- Skapar matematisk modell
- Exempel:
  - Linjär regression
  - Logistic regression
  - Neurala nätverk

---

## **Kombinationsprincip**

Varje algoritm kan placeras i alla fyra dimensioner.

Exempel:

**Linjär regression**
- Supervised
- Regression
- Model-Based
- Batch

**Logistic Regression**
- Supervised
- Klassificering
- Model-Based
- Batch

**K-Means**
- Unsupervised
- Klustring
- Model-Based

**KNN**
- Supervised
- Klassificering / Regression
- Instance-Based
- Batch

**Neurala nätverk**
- Supervised / Self-Supervised
- Regression / Klassificering
- Model-Based
- Batch (oftast)

**Reinforcement Learning + Neural Networks**
- Reinforcement
- Online
- Model-Based
