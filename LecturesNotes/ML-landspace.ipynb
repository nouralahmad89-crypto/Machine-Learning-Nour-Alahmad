# **Machine Learning Landscape**

Maskininlärning kan delas in i fyra huvuddimensioner.  
När du lär dig en algoritm ska du kunna placera den i varje kategori.

---

## **Inlärningsparadigm (Hur lär sig modellen?)**

### 🔵 Supervised Learning
- Har etiketter (facit)
- Modellen lär sig genom att minimera fel
- Exempel:
  - Huspris (Regression)
  - Spam/inte spam (Klassificering)
  - Kreditrisk

---

### 🟡 Unsupervised Learning
- Inga etiketter
- Hittar struktur i data
- Exempel:
  - Klustring
  - Segmentering
  - Kundgrupper

---

### 🟠 Semi-Supervised Learning
- Lite etiketter
- Mycket oetiketterad data
- Används när etiketter är dyra

---

### 🟣 Self-Supervised Learning
- Maskar delar av input
- Förutsäger saknade delar
- Exempel:
  - GPT (maskar ord)
  - BERT
  - Bildmodeller

---

### 🔴 Reinforcement Learning
- Lär sig via belöning
- Interagerar med miljö
- Action → Reward → Uppdatering
- Exempel:
  - Spel-AI
  - Robotik
  - Trading

---

## **Typ av Uppgift (Task)**

### Regression
- Förutsäger numeriskt värde
- Exempel:
  - Pris
  - Temperatur
  - Tid

---

###  Klassificering
- Förutsäger kategori
- Exempel:
  - Spam
  - Sjukdom
  - Bildklassificering

---

###  Klustring
- Hittar grupper
- Ofta Unsupervised

---

### Dimensionality Reduction
- Minskar antal features
- Exempel:
  - PCA
  - Kompression

---

###  Anomaly Detection
- Hittar ovanliga mönster
- Exempel:
  - Kreditkortsbedrägeri

---

###  Novelty Detection
- Hittar något nytt (positivt ovanligt)

---

## **Träningssätt**

### 🔵 Batch Learning
- Tränas en gång
- Deployas
- Uppdateras inte kontinuerligt

---

### 🟡 Online Learning
- Uppdateras kontinuerligt
- Exempel:
  - Rekommendationssystem
  - Realtidsdata

---

##  **Modelltyp**

### 🔵 Instance-Based
- Lagrar data
- Jämför nya datapunkter med gamla
- Ingen riktig generalisering
- Exempel:
  - KNN

---

### 🟡 Model-Based
- Lär parametrar
- Skapar matematisk modell
- Exempel:
  - Linjär regression
  - Logistic regression
  - Neurala nätverk

---

## **Kombinationsprincip**

Varje algoritm kan placeras i alla fyra dimensioner.

Exempel:

**Linjär regression**
- Supervised
- Regression
- Model-Based
- Batch

**Logistic Regression**
- Supervised
- Klassificering
- Model-Based
- Batch

**K-Means**
- Unsupervised
- Klustring
- Model-Based

**KNN**
- Supervised
- Klassificering / Regression
- Instance-Based
- Batch

**Neurala nätverk**
- Supervised / Self-Supervised
- Regression / Klassificering
- Model-Based
- Batch (oftast)

**Reinforcement Learning + Neural Networks**
- Reinforcement
- Online
- Model-Based
