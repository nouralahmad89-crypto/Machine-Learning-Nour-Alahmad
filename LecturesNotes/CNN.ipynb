## Convolutional Neural Network:

Vi använder CNN när vi har bilddata och vill att modellen ska känna igen mönster och klassificera dem.

**Convolution :** små filter (kernels) glider över bilden och hittar mönster.

Stor bild → delas upp i små delar → analyseras → sätts ihop igen.

En kärna (kernel) 3×3 eller 4×4 som vandrar över bilden.

**Vad är en kernel?**
- En liten matris (t.ex. 3×3)
- Innehåller vikter (siffror)
- Den används för att analysera en liten del av bilden

Kernel flyttas pixel för pixel över hela bilden ,På varje position gör den en beräkning.

- Ta en liten del av bilden (t.ex. 3×3)
- Multiplicera med kernel
- Summera allt ,→ få 1 värde (pixel) i output.

Resultat :En ny bild skapas → kallas feature map . Den visar:vad filtret “ser” i bilden.

Olika kernels lär sig:

- kanter
- färger
- former

**Hur används bias?**

För varje pixel i en feature map: `output = (input * weights) + bias`, samma bias används över hela feature mapen. -------> `output = convolution + bias`

`padding="same"` Utan padding: kernel kan inte gå utanför bilden → bilden blir mindre

Med padding:vi “lägger en ram av nollor” runt bilden → kernel kan täcka hela bilden
→ storleken bevaras, **bra och enkelt (speciellt vid stride=1)**

`padding=...` bättre när du vill ha kontroll

Stride = hur många pixlar kernel hoppar varje gång.

Större stride → mindre output-bild , Mindre stride → mer detaljer.

vi minskar storleken på bilden ganska drastiskt” “padding bevarar storleken .**Padding =** lägga till extra pixlar runt bilden ( ofta nollor).

CNN lär sig i flera nivåer (lager). den bygger upp förståelsen steg för steg.

först enkla mönster (kanter), sedan former, och till slut hela objekt som t.ex. hund eller katt.

**Pooling** minskar bildens storlek genom att sammanfatta områden (t.ex. ta max eller medelvärde).

Två typer :

- Max Pooling :tar största värdet ,behåller starkaste signal.

- Average Pooling :tar medelvärde , mer “smooth” .

pooling = mindre tensor → nästa layer.

**Depth-wise max pooling** är en metod där man gör pooling över kanaler (feature maps) istället för över bildens höjd och bredd.

Hur det fungerar:

- Vi tar flera feature maps (kanaler)
- jämför deras värden position för position
- väljer det största värdet (max)
- skapar en ny feature map

CNN bygger förståelse genom att upprepa: convolution → aktivering → pooling flera gånger.

Varje block gör:

1. Convolution → hittar mönster
2. ReLU → gör det icke-linjärt
3. Pooling → minskar storlek

Flatten omvandlar feature maps till en vektor så att den kan användas av ett vanligt neuralt nätverk.

Fully Connected-lagret använder alla features för att fatta ett slutgiltigt beslut (klass).

`nn.Dropout(0.5)` Dropout stänger slumpmässigt av neuroner under träning för att göra modellen mer robust.
50% av neuronerna stängs av varje gång ,olika neuroner varje gång för att undvika överanpassning (overfitting).

`model.train()` -> dropout aktiv . `model.eval()` dropout AV.

Idéen är att varje lager minskar upplösningen och mängden detaljer i bilden och därmed lär sig känna igen mer och mer abstrakta detaljer.

Dropout är en regulariseringsteknik inom maskininlärning som används för att minska överanpassning (overfitting) i neurala nätverk.

`nn.Dropout(0.5)` 50% chans att en nod inte används under träning.

`model.eval()` stänger av dropout. (“jag testar det jag lärt mig”)

`model.train()` → “jag lär mig” 

**CNN-del:**

extraherar features (kanter → former → objektdelar)

**Klassificeringsdel:**

tar features → gör beslut (klass)

- Varje pixel bidrar som input (som en nod)
- Kernel rör sig över bilden istället för full koppling
- Samma vikter används över hela bilden
