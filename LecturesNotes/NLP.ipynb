- Ta bort symboler

MEN: Varje steg = kompromiss mellan enkelhet och information.

### **Olika typer av tokenization (hur vi delar upp text)**

1. Word-level tokenization (ordnivå):

Fördelar: Enkelt , Lätt att förstå

Problem: Explosion av antal ord (vocabulary) : Det blir för många tokens och svårt att hantera.

2. Subword tokenization (väldigt viktigt!): Här delar man ord i delar

Fördel: Mindre vocabulary , Klarar nya ord bättre . Detta är väldigt användbart i praktiken.

Speciellt viktigt för språk som svenska (sammansatta ord) , andra “komponerande språk”

3. Character-level (tecken-nivå) : Här är varje bokstav en token

Fördel: Klarar ALLA språk ,Inga “okända ord”

Problem: För mycket detaljer , svårt att förstå meaning ,längre sekvenser

4. N-grams: Här tittar man på flera ord tillsammans

Man går sällan över 3-gram . Varför? Kombinationer exploderar 💥 , för mycket data.

Valet påverkar: modellens kvalitet , komplexitet , minne

**_Other tokenizations_**

1. MORFEM = minsta betydelsebitar

Exempel:
"unhappy"
→ ["un", "happy"]

2) LEXEM = samma ord, olika former

Exempel:
run, runs, running
→ samma lexem: RUN

MEN kontext är viktigt:

"take care"
"take over"
"I don't care"

→ samma ord, olika betydelser

--------------------------------------

3) GRAFEM = hur det skrivs

Exempel:
"ch" → ett ljud men två bokstäver

Språkexempel:
Franska → många bokstäver uttalas inte
Irländska → komplex stavning

Slutsats:
det vi SER ≠ det vi hör

--------------------------------------

4) FONEM = ljud

Exempel:
"bat" vs "pat"
→ ett ljud ändras → ny betydelse

Svenskt exempel:
"tomten" (Santa)
"tomten" (mark)

→ samma stavning, olika ljud/betydelse

--------------------------------------
SLUTSATS:

Perfekt språkförståelse = svårt
Praktisk lösning = enklare representation (tokens)

### **Embeddings (hur ord får meaning)**

PROBLEM:
Index (1,2,3) har ingen meaning

LÖSNING:
Embeddings = vektorer

"I" → [0.1, 0.8, 0.3]

VIKTIGT:
Liknande ord → nära varandra

HUR:
Tränas på stora textmängder

RESULTAT:
Modellen lär sig meaning

MEN:
Kan innehålla bias från data

1. Text → tokens
"I love AI" → ["I", "love", "AI"]

2. Token → embedding (vektor)
"I" → [0.1, 0.8, 0.3]

3. Dessa vektorer → input till neuralt nätverk

Embedding-vektorerna ÄR vikter som lärs under träning. varje token har redan en inlärd vektor (vikter).Neuralt nätverk använder dessa för att: förstå samband , förutsäga nästa ord.

Embeddings lärs AUTOMATISKT baserat på hur ord används i text.

Under training the model learns embeddings and weights by minimizing prediction errors.
During inference it uses these learned representations to process new input.

Under träning uppdateras alla embeddings samtidigt,
och de påverkar varandra genom kontext.

1. Embeddings startar slumpmässigt
2. Modellen gör prediction
3. Räknar ut fel (loss)
4. Uppdaterar:
   - embeddings
   - andra vikter i nätverket
5. Upprepas många gånger

Embedding = one-hot × vikt-matris

under träning :

- Vikt-matrisen W uppdateras
- → embeddings förändras
- → meaning skapas

`Token → ID → Embedding → Neural Network (MLP / Transformer)`

`one-hot → embedding (linear layer)  vektor → MLP (eller annan modell)`

Det finns flera pipelines:

1. TF-IDF → cosine (enkelt)
2. TF-IDF → NN (mer kraftfullt)
3. Embedding → NN (modern standard)

### **TF-IDF vs Embeddings + effektivitet**

TF-IDF:
- baserat på frekvens
- ingen meaning
- sparse
- enkel

Embeddings:
- lär meaning
- dense
- kompakt
- mer kraftfull

TF-IDF → sparse (mest nollor)

Embeddings → dense (kompakt info)

**Två olika sätt att använda TF-IDF**

1. _Klassisk metod (utan NN):_ TF-IDF → cosine similarity → jämförelse , (regelbaserat
ingen träning)

används för: likhet mellan texter , sökning (search engines)

2. _Med neuralt nätverk :_ TF-IDF → MLP (NN) → klassificering (lär sig från data
kan hitta komplexa mönster)

används för: spam detection , sentiment analysis , textklassificering

### **Träningskostnad + problem (alignment & safety)**

Training:
- extremt dyrt
- kräver stora resurser

Pretraining:
- tränas på internetdata

Problem:
- bias
- fel associationer
- konstiga svar

Alignment:
- modellen speglar data

Safety:
- modellen kan lära farliga beteenden

`CNN: lika bilder → lika features , NLP: lika ord → lika vektorer`

**Alignment** handlar om att modellen lär rätt saker från rätt data för rätt uppgift.(rätt värderingar)

Alignment refers to ensuring that learned representations reflect appropriate and accurate relationships for the intended task. Poor alignment occurs when models learn biases or irrelevant patterns from training data, leading to incorrect or misleading associations.

**Safety =** att modellen inte gör farliga eller skadliga saker (undvika skada)
