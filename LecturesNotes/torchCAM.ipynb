### **Vad gör TorchCAM?**

TorchCAM hjälper att svara på: Vilka delar av bilden använde modellen för sitt beslut?

Den gör detta genom att skapa en activation map (heatmap).

### **Hur det funkar (enkelt)**

TorchCAM (t.ex. LayerCAM) gör:

- Tar output från modellen
- Kollar vilken klass som är viktigast (argmax)
- Spårar tillbaka till ett convolution layer
- Ser vilka pixlar som bidrog mest
- Skapar en heatmap

Det är en typ av feature attribution. Vi kan säga att CNN lär sig features (kanter → former → objekt)
TorchCAM visar vilka features användes.

Det här är inte perfekt:
- modellen kan ha rätt svar men fel fokus ❌
- modellen kan ha fel svar men rimlig fokus 🤔

Hela flödet:
- Läs bild → (C, H, W)
- preprocess → (C, H, W)
- unsqueeze → (1, C, H, W)
- model → output
- squeeze → ta bort batch

Pretrained models kräver samma preprocessing som de tränades med. `weights.transforms()` används för att säkerställa att input preprocessas på samma sätt som under träningen av modellen.

### **Vad är LayerCAM?**

En metod för att skapa attribution maps.Den svarar på: Vilka pixlar bidrog mest till beslutet?
