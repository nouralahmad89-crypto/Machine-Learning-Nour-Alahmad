## 🔹 1. Bag of Words (BoW)

👉 Representerar text som antal ord

### Exempel:
Doc1: "this movie is good"  
Doc2: "this movie is bad"

Vocabulary:
[this, movie, is, good, bad]

BoW:

|      | this | movie | is | good | bad |
|------|------|-------|----|------|-----|
| Doc1 | 1    | 1     | 1  | 1    | 0   |
| Doc2 | 1    | 1     | 1  | 0    | 1   |

✔ Enkelt  
❌ Alla ord lika viktiga  

---

## 🔹 2. Term Frequency (TF)

👉 Hur ofta ett ord finns i dokumentet

TF(t,d) = (antal gånger ordet) / (totalt antal ord)

### Exempel:
"good" i Doc1 → 1/4 = 0.25

✔ Normaliserar längd  
❌ Vanliga ord fortfarande viktiga  

---

## 🔹 3. TF-IDF

👉 Vägt mått på ordets betydelse

TFIDF(t,d) = TF(t,d) × IDF(t)

IDF(t) = log(N / df)

- N = antal dokument  
- df = antal dokument med ordet  

---

### Intuition:

| ord   | TF | IDF | TF-IDF |
|-------|----|-----|--------|
| this  | hög | låg | låg |
| good  | hög | hög | ⭐ hög |

✔ Viktiga ord får hög vikt  
✔ Vanliga ord får låg vikt  

---

## 🔹 4. Cosine Similarity

👉 Mäter likhet mellan två dokument

cos(θ) = (A · B) / (||A|| ||B||)

| värde | betydelse |
|------|----------|
| 1    | lika |
| 0    | olika |

✔ Bryr sig om riktning, inte längd  

---

## 🔹 Pipeline (viktig!)

text → BoW → TF → TF-IDF → cosine similarity

---

# ✅ Minnesregler

- BoW = räkna ord  
- TF = hur ofta  
- IDF = hur ovanligt  
- TF-IDF = vikt  
- Cosine = likhet  

Cosine similarity is used in NLP **( Natural Language Process)** to measure how similar two documents are by comparing their vector representations.

Each document is converted into a vector (e.g., using Bag-of-Words or TF-IDF), and cosine similarity measures the angle between these vectors.

Using Bag-of-Words, similarity is based on raw word counts, so common words can dominate the result.

Using TF-IDF, words are weighted by importance, giving less weight to common words and more weight to unique ones. Therefore, cosine similarity with TF-IDF provides a more meaningful measure of document similarity.
