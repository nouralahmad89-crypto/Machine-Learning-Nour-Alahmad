Ett nätverk = massa multiplikationer + funktioner.

`(W·X + b) → activation → nästa lager → output`

Så funkar en neuron (nod):

För varje nod:

- Tar input (data)
- Multiplicerar med vikter (weights)
- Lägger till bias
- Skickar genom en funktion

Detta är egentligen bara **matrismultiplikation**

Bias : Gör modellen mer flexibel och flyttar resultatet upp/ner.

Ett nätverk består av lager (layers) , Varje lager = massa noder

Input → Layer → Layer → Output

Med **aktiveringsfunktion**: Nätverket kan lära sig komplexa mönster ,icke-linjära samband. Detta gör modellen icke-linjär och mycket kraftfullar.

Ett lager = en matris

Flera lager = flera matriser

Hela nätverket = tensor (flerdimensionell matris).


### Activation Functions

An activation function is applied to determine what information is passed to the next layer.

### Vanliga typer

#### ReLU (Rectified Linear Unit)
- f(x) = 0 om x < 0, annars x
- Mest använda funktionen

**Fördelar:**
- Snabb och enkel
- Fungerar bra i djupa nätverk

**Nackdelar:**
- Kan orsaka "döda neuroner"

---

#### Sigmoid
- Output mellan 0 och 1

**Används för:**
- Sannolikheter
- Binär klassificering

**Nackdelar:**
- Vanishing gradient-problem

---

#### Tanh
- Output mellan -1 och 1

**Fördelar:**
- Centrerad runt 0
- Ofta bättre än sigmoid

---

### Tumregel
- ReLU → dolda lager  
- Sigmoid → output (klassificering)  
- Tanh → alternativ till sigmoid  

---

### Modellträning

Målet att justera vikter (W) för att minimera felet.

#### **Steg**

1. **Forward pass :** (beräkna output)

- Input skickas genom nätverket
- Beräkna prediktion

2. **Beräkna förlust (loss):** (jämför med true value)

- Jämför prediktion med verkligt värde.

3. **Backpropagation:** (räkna gradienter)

- Beräkna gradienter (hur vikterna påverkar felet)

4. **Gradient descent:** (justera vikter)

- Uppdatera vikterna. `W = W - learning_rate * gradient`

5. **Upprepa:** (förbättra modellen)

- Kör flera iterationer (epochs)


#### Intuition
- Modellen gör en prediktion  
- Mäter felet  
- Justerar vikterna  
- Förbättras steg för steg  

**Neuralt nätverk =**  
(W·X + b) → aktivering → uppdateras med gradient descent.

- Backpropagation räknar gradienter
- Chain rule kopplar ihop lagren
- Gradient descent uppdaterar vikter
- Upprepning → bättre modell
