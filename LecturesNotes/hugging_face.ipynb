### **Hugging Face :**

En plattform där man kan:

- ladda ner färdiga AI-modeller
- hitta dataset
- testa modeller direkt
- använda dem i Python

###  **Basmodell vs Fine-tuned modell**

Huvudidé : Skillnaden är om modellen är tränad för att **användas direkt** eller inte.

#### Basmodell (Rå modell)
- Tränad på mycket text  
- Förstår språk  

Men: följer inte instruktioner bra  

#### Fine-tuned (IT) (Redo att använda)
- Basmodell + extra träning  
- Kan:
  - svara på frågor  
  - följa instruktioner  

**IT = Instruction Tuned**
I praktiken används nästan alltid _fine-tuned modeller_

#### **attention_mask**

Vad är det?
- 1 = riktig token
- 0 = padding (<pad>)

Syfte : ignorera padding i modellen

Resultat : samma shape som input


Padding och mask fungerar tillsammans:

- padding → gör data rätt form
- mask → ser till att modellen inte blir lurad

#### **Positional Encodings**

Problem : modellen vet inte ordning

Lösning : lägg till position embedding

Resultat : embedding = token + position

Viktigt : ordning påverkar betydelse . Insikt : modellen får både vad och var

#### **Transformer:**

Transformers gör något smart:

- De tittar på alla ord samtidigt
- De lär sig relationer mellan ord direkt

**Attention =** modellen bestämmer vilka ord den ska fokusera på

- Q = Query (fråga)
- K = Key (nyckel)
- V = Value (värde)

**Self-attention =** varje ord gör attention mot ALLA ord i samma mening

- Tokenizer → “vad är orden?”
- Embedding → “vad betyder orden?”
- Position → “var står orden?”
- Attention → “hur hänger orden ihop?

**hur en Transformer genererar text steg för steg?**

1. Tokenizer : Text delas upp i tokens

"I love AI" → ["I", "love", "AI"]

2. Embedding : Varje token → vektor (siffror)

"love" → [0.2, -0.5, 1.0] → representerar betydelse

3. Position Encoding : Lägger till position

final vector = embedding + position ( viktigt för ordning)

4. Encoder (förstå input)
- använder self-attention  
- varje ord tittar på alla andra  

→ output: kontext-rika vektorer

5. Decoder (börjar generera)
Input till decoder:
- encoder output  
- tidigare genererade ord  

6. Förutsäga nästa token
Decoder använder:
- input + tidigare ord  

→ producerar en vektor (hidden state)

7. Output layer (logits)

logits = h · W + b   → poäng för varje ord


8. Softmax (sannolikheter)

P(y_i) = e^{z_i} / Σ e^{z_j}

→ gör logits till sannolikheter

9. Välj nästa ord

"Jag älskar" → "AI"

10. Upprepa

<start>
→ Jag
→ Jag älskar
→ Jag älskar AI
→ <end>

#### **Nyckelidé**

- Encoder = förstå  
- Decoder = generera  
- Attention = koppla ord  
- Softmax = välja nästa ord.

Transformer läser en mening, förstår relationer, och genererar en ny mening ett ord i taget.
