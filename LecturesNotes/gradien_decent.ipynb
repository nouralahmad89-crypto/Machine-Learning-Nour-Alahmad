# Gradient Descent 

## Vad är Gradient Descent?
Gradient Descent är en **optimeringsalgoritm** som används i maskininlärning för att **minimera en loss-funktion** och hitta de bästa modellparametrarna.

## Grundidé
- Beräkna gradienten (lutningen) av loss-funktionen
- Gå ett steg i **motsatt riktning mot gradienten**
- Upprepa tills felet slutar minska

Gradient = hur mycket felet ändras när vi ändrar parametrarna.

## Formel
θ(t+1) = θ(t) − η ∇L(θ)

- θ = modellparametrar  
- η = learning rate (steglängd)  
- ∇L(θ) = gradient av loss-funktionen

## Learning Rate
Bestämmer hur stort steg modellen tar.

- För liten → långsam träning, kan missa optimum  
- För stor → kan hoppa över minimum och bli instabil

## Stop Condition
Gradient Descent stoppar inte automatiskt. Vanliga stoppvillkor:
- fast antal **epochs**
- **early stopping** (loss förbättras inte längre)

## Epoch
En **epoch** = en full genomgång av hela datasetet.

## Typer av Gradient Descent
1. **Batch Gradient Descent**  
   - använder hela datasetet  
   - stabil men långsam

2. **Stochastic Gradient Descent (SGD)**  
   - använder en datapunkt i taget  
   - snabb men instabil

3. **Mini-Batch Gradient Descent**  
   - använder små grupper av datapunkter  
   - mest använd i praktiken

## Förbättringar
**Momentum / Adam**
- adaptiv learning rate
- hjälper modellen att undvika lokala minimum

# **Derivata**

## Vad är derivata?
Derivata beskriver **hur snabbt en funktion förändras när en variabel ändras**.  
Den kan tolkas som **lutningen på en kurva**.

I machine learning används derivator för att **minimera en loss-funktion**.

---

## Gradient Descent
Parametrar uppdateras med derivatan:

θ = θ − η ∇L(θ)

- θ = modellparametrar  
- η = learning rate  
- ∇L(θ) = gradient (derivatan av loss)

Gradienten visar **riktningen där felet ökar mest**, därför går vi **åt motsatt håll**.

---

## Viktiga deriveringsregler

**Konstant**
f(x) = c  
f'(x) = 0

**Potensregel**
f(x) = xⁿ  
f'(x) = n xⁿ⁻¹

**Sumregel**
(f + g)' = f' + g'

**Konstant × funktion**
(c f(x))' = c f'(x)

**Kedjeregel**
Om funktioner är inuti varandra:  
f(x) = (g(x))ⁿ  
f'(x) = n(g(x))ⁿ⁻¹ · g'(x)

---

## Exempel i ML (MSE)

Loss:

MSE(θ) = (1/n) Σ (y − ŷ)²

Gradient:

∇MSE(θ) = (2/n) Xᵀ(Xθ − y)

Används i Gradient Descent:

θ ← θ − η (2/n) Xᵀ(Xθ − y)

---

## Intuition

Derivatan svarar på:

> Hur ska parametrarna ändras för att minska modellens fel?
