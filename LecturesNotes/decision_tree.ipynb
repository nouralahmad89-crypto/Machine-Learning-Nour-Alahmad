### 🌳 Decision Trees – kort sammanfattning


Ett decision tree är en modell som fattar beslut genom att ställa frågor i flera steg

* **Vad det är:**
  En modell som delar upp data steg för steg genom regler (t.ex. “om petal length < 2 → gå vänster”). Varje slutpunkt (leaf) ger en klass eller ett värde.

* **Hur den fungerar:**

  * Börjar i roten och gör uppdelningar (splits) baserat på features
  * Målet är att göra grupper så “rena” som möjligt (samma klass i varje grupp)
  * Använder mått som **Gini impurity eller entropi** för att välja bästa split

* **Viktiga egenskaper:**

  * Icke-linjär modell (kan hantera komplexa samband)
  * Lätt att visualisera och tolka (white-box modell)
  * Kräver inte skalning av data

* **Problem:**

  * Kan lätt bli **överanpassad (overfitting)** om trädet blir för djupt
  * Små förändringar i data kan ge olika träd

* **Hur man motverkar overfitting:**

  Man begränsar trädet:

- max_depth → hur djupt trädet får bli

- min_samples_split → min antal datapunkter för att dela

- min_samples_leaf → min antal i en leaf

- pruning → (ccp_alpha) → klippa bort delar

- max_features -> hur många features som testas vid varje split
- max_leaf_nodes -> max antal slutnoder
- random_state -> gör resultatet reproducerbart

* **Extra:**

Vad betyder min_samples_leaf=5?

Varje leaf måste innehålla minst 5 datapunkter.

Så trädet får inte skapa väldigt små slutnoder.

Varför är det bra?

För att små leaves ofta betyder att trädet:

- fångar brus
- memorerar träningen
- överfittar
  * Decision trees används också i **ensemble-metoder** som Random Forest
  * Feature engineering är fortfarande viktigt för bra resultat

**Vad modellen försöker minimera**

- I klassificering:👉 minskar Gini / Entropy

- I regression: 👉 minskar MSE (Mean Squared Error)
