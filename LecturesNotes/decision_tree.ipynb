### 🌳 Decision Trees – kort sammanfattning


Ett decision tree är en modell som fattar beslut genom att ställa frågor i flera steg

* **Vad det är:**
  En modell som delar upp data steg för steg genom regler (t.ex. “om petal length < 2 → gå vänster”). Varje slutpunkt (leaf) ger en klass eller ett värde.

* **Hur den fungerar:**

  * Börjar i roten och gör uppdelningar (splits) baserat på features
  * Målet är att göra grupper så “rena” som möjligt (samma klass i varje grupp)
  * Använder mått som **Gini impurity eller entropi** för att välja bästa split

* **Viktiga egenskaper:**

  * Icke-linjär modell (kan hantera komplexa samband)
  * Lätt att visualisera och tolka (white-box modell)
  * Kräver inte skalning av data

* **Problem:**

  * Kan lätt bli **överanpassad (overfitting)** om trädet blir för djupt
  * Små förändringar i data kan ge olika träd

* **Hur man motverkar overfitting:**

  Man begränsar trädet:

- max_depth → hur djupt trädet får bli

- min_samples_split → min antal datapunkter för att dela

- min_samples_leaf → min antal i en leaf

- pruning → (ccp_alpha) → klippa bort delar

- max_features -> hur många features som testas vid varje split
- max_leaf_nodes -> max antal slutnoder
- random_state -> gör resultatet reproducerbart

* **Extra:**

Vad betyder min_samples_leaf=5?

Varje leaf måste innehålla minst 5 datapunkter.

Så trädet får inte skapa väldigt små slutnoder.

Varför är det bra?

För att små leaves ofta betyder att trädet:

- fångar brus
- memorerar träningen
- överfittar
  * Decision trees används också i **ensemble-metoder** som Random Forest
  * Feature engineering är fortfarande viktigt för bra resultat

**Vad modellen försöker minimera**

- I klassificering:👉 minskar Gini / Entropy

- I regression: 👉 minskar MSE (Mean Squared Error)


### Decision Tree för regression :

för regression tar man medelvärdet av alla datapunkter som hamnar i en nod/klass/regional del av trädet.

**Vad betyder det?**

- När ett decision tree används för klassificering:

modellen säger vilken klass en punkt tillhör

- När ett decision tree används för regression:

modellen säger ett tal
det talet är oftast medelvärdet av target-värdena för träningspunkterna som hamnade i samma blad.

varje slutregion får ett värde som är medelvärdet av alla y-värden i den regionen.

Decision Tree Regressor fungerar så här:

1. dela upp feature-rummet med splits
2. skapa blad
3. i varje blad: ta medelvärdet av target-värdena
4. använd det medelvärdet som prediktion

👉 Decision trees överfittar snabbt eftersom de kan dela upp datan tills de i princip memorerar varje datapunkt.

`tree_reg.tree_.threshold`
`tree_reg2.tree_.threshold`
Det här visar själva splitpunkterna.

Alltså de x-värden där trädet säger:

- vänster gren om mindre än tröskeln
- höger gren om större än tröskeln

Så de här arrayerna visar exakt var trädet skär upp x-axeln.

Decision trees har hög varians eftersom:

1. första splitten är instabil
2. små dataskillnader → stora modellskillnader
3. greedy beslut (lokala)
4. inga mjuka övergångar
5. dominoeffekt i hela trädet

inte bara på själva datan

istället för: x1, x2 får du: nya axlar som följer datans riktning

man måste standardisera innan PCA

### Ensemble Learning (lösningen på hög varians)

Använd många modeller istället för en, många träd → stabilare resultat

Vi tar många träd och medelvärdesbildar deras prediktioner:

slumpen jämnas ut
modellen blir stabil

Nu kommer “hur vi kombinerar modellerna”---> **Voting** 

1. hard voting : varje modell röstar på en klass--> majoritet
2. soft voting : vi använder --> medel av sannolikheter

soft voting ger ofta bättre resultat
### Bagging

vi tränar samma modell många gånger på lite olika data

🔹 Hur det funkar
1. ta dataset
2. skapa många slumpmässiga kopior (med återläggning) i.e (samma datapunkt kan väljas flera gånger)
3. träna ett träd per dataset
4. låt dem rösta

#### Out-of-bag

vissa datapunkter används inte i träningen , dessa kallas out-of-bag data

De användas som testdata så ,kan mäta hur bra modellen är utan att ha ett separat test-set.

Bagging → med återläggning
Pasting → utan återläggning

### Random Forest (förbättrad bagging)

random forest = många beslutsträd + extra slump ,  Random Forest = Bagging + slumpmässiga features

Hur den funkar
1. ta slumpmässiga datapunkter (bagging)
2. träna många träd
3. varje split använder slumpmässiga features
4. kombinera:
- klassificering → röstning
- regression → medelvärde

kan ge feature importance

#### Boosting

träd som lär sig från varandra. Träna modeller i serie ,varje ny modell lär sig från tidigare fel. 

modellen lär sig från sina fel istället för träna alla träd samtidigt gör vi:

träna ett träd i taget ,varje nytt träd fokuserar på felen.

**AdaBoost**

- ger mer vikt till fel datapunkter
- nästa modell fokuserar på dem

💡 “svåra punkter blir viktigare

**Gradient Boosting**

- tränar på residualer (fel)
- varje modell förbättrar föregående

💡 “lär sig felen direkt” (minskar bias)

#### Stacking :

istället för att bara rösta → vi tränar en modell som kombinerar de andra.

istället för en enkel funktion (soft voting)
tränar vi en modell som gör kombinationen

👉 alltså:

- voting = fast regel
- stacking = lärd regel
