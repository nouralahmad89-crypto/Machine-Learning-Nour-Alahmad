## **_PCA_** 

Är en metod för dimensionsreduktion ,  dimensionsreduktion = minska antal features (kolumner).

_Problem:_

- för många dimensioner → svårt för modellen
- data blir gles
- risk för overfitting

_Lösning:_

PCA förenklar datan utan att tappa för mycket information

### **Problemet – Curse of Dimensionality**

- varje feature har olika skala
- mycket av rummet är tomt
- punkter ligger långt ifrån varandra
-  modellen blir dålig

==> detta är curse of dimensionality

**Lösningar**

- Skala data (StandardScaler)
- Samla mer data
-  Minska dimensioner → PCA (bästa lösningen här)

### **Hur PCA fungerar (Principal Components + varians)**

PCA skapar nya features som kallas: Principal Components (PCs).

Vad är en “principal component”?

En principal component är: en linjär kombination av alla dina features ,som fångar maximalt med varians

Första komponenten (PC1) fångar mest varians , Andra komponenten (PC2) fångar näst mest , är ortogonal (vinkelrät) mot PC1    osv…

Features blir:oberoende -> inte korrelerade . detta löser: multikollinearitet

#### **Vad PCA egentligen gör visuellt?**

roterar koordinatsystemet , lägger nya axlar där datan sprids mest

**Behåll så få dimensioner som möjligt – utan att tappa för mycket varians**

PCA handlar inte bara om accuracy

Den hjälper också med:

- snabbare träning
-  mindre minne
- stabilare modell
