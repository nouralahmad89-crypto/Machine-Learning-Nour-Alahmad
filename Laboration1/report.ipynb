

### **Content-Based Recommendation with TF-IDF and Cosine Similarity**

- Course  : Machine Learning – Assignment 1
- Dataset : MovieLens (`movies.csv`, `tags.csv`)

#### _1. Objective_

The goal of this project is to build a content-based movie recommendation system that suggests similar movies based on their content.

The system recommends top-N movies given an input movie by analyzing:

- genres
- user-generated tags


#### _2. Method Overview_

The recommendation system follows these steps:

1. Combine movie genres and tags into a single text feature (content)
2. Convert text into numerical vectors using TF-IDF
3. Compute similarity between movies using cosine similarity
4. Return the most similar movies

#### _3. Data Preparation_

- Movies with missing or irrelevant genres were removed
- Tags were aggregated per movie
- Text was cleaned (lowercase, remove duplicates, handle missing values)
- A new column content was created ( genres + tags).

#### _4. Feature Extraction (TF-IDF)_

TF-IDF was used to convert text into vectors:
This method:

- highlights important words in each movie
- reduces the impact of common words
Each movie is represented as a numerical vector

#### _5. Similarity Computation_

Similarity between movies is calculated using cosine similarity:
- Values range from 0 to 1
- Higher values indicate more similar movies

#### _6. Recommendation Process_

1. Convert input movie title to index
2. Compute similarity scores
3. Sort movies by similarity
4. Exclude the input movie
5. Return the top-N recommendations

A content-based recommendation system was successfully implemented using TF-IDF and cosine similarity.
The system can recommend movies with similar content, but future improvements could include collaborative filtering for personalization.
