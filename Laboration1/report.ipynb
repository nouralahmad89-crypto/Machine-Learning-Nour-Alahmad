## 1. Assignment context

This notebook documents the implementation of a **content-based movie recommendation system** using the MovieLens dataset.  
The assignment asks for a system that can recommend **five movies from an input movie**, and it explicitly allows both **content filtering** and **collaborative filtering**. It also mentions that combining **genres** and **tags** with **TF-IDF vectorization** is a strong approach for capturing deeper movie similarity.

The current notebook focuses on the **content-based stage** of the project.  
The next planned stage is to extend the work with **collaborative recommendation** based on user ratings.

---

## 2. Goal

The goal of this stage is to recommend movies that are similar in content to a selected movie by:

1. Using **genres** and **user-generated tags** as textual movie features  
2. Cleaning and aggregating the metadata  
3. Converting movie content into TF-IDF vectors  
4. Comparing movies with **cosine similarity**  
5. Returning the **top-N most similar movies**
